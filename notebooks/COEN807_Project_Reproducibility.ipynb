## Dataset and target engineering

The project uses the Portuguese-language subset of the UCI Student Performance dataset. The binary target is `At_Risk = 1` when `G3 < 10`, otherwise `0`.

```python
students = pd.read_csv(DATA_PATH, sep=';')
students['At_Risk'] = (students['G3'] < 10).astype(int)
print('Rows:', len(students))
print('Original predictors:', students.shape[1] - 1)
students['At_Risk'].value_counts().rename(index={0: 'Not at risk', 1: 'At risk'})
```

## Saved experimental comparison

The table below contains the metrics generated by the two experiments: early warning without `G1` and `G2`, and performance aware with `G1` and `G2`.

```python
metrics = pd.read_csv(RESULTS_PATH)
metrics.sort_values(['experiment', 'f1_at_risk'], ascending=[True, False])
```
